Plotting the data using k modes as the data is mostly categorical

```r
library(klaR)
```

```r
x <- data1[,c("sex","age","inc","race","party")]
x <- x[complete.cases(x), ]
cl_3 <- kmodes(x, 3)
```

```r
cl_3
```

---------------3 clusters--------------------------
The largest cluster has mostly people with age around 72 and females whose income lie between 50k to 70k. These women mostly have democratic views.
Hence analyzing the group members for this cluster it can be inferred that 70 plus females would mostly be democratic in view.
Also the other cluster is fairly large enough and its group members are males whose income is quite large, above 150k and they are independent in their political 
views. This cluster hence can be used to infer that mostly succsesful men are independent
in their political views.
The third cluster is much smaller than other two and it has women with much 
less income, around 10k to 20k and age around 44.

```r
cl_4 <- kmodes(x, 4)
```

```r
cl_4
```

Taking 4 clusters gives a wider overview of the scenario.
However here it can be seen that women with age around 72 have republic views.
They have income around 50k to 70k and repulican political views.
The cluster is almost the same size as was the largest cluster in 3 clustering process.
But here the same type of group members have republican views.
However taking larger number of clusters gives an insight into the younger generation
who have mid way income and democratic views
